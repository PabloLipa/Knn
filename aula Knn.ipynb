Importando bibliotecas

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import make_blobs
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
```

criar dataset

```python
X, y = make_blobs(n_samples= 500, n_features=2, centers=4, cluster_std= 1.5, random_state=42)
```

vizualizar dataset

```python
plt.style.use('seaborn')
plt.figure(figsize=(10,10))
plt.scatter(X[:,0],X[:,1], c=y, marker='*',s=100,edgecolors='black')
plt.show()
```

treino e teste! holdout

```python
X_train, X_test, y_train, y_test  = train_test_split(X,y, random_state=42)
```

K-nn

```python
knn5 = KNeighborsClassifier(n_neighbors=5)
knn3 = KNeighborsClassifier(n_neighbors=3)
```

predio

```python
knn5.fit(X_train,y_train)
knn3.fit(X_train,y_train)

y_predicao5 = knn5.predict(X_test)
y_predicao3 = knn3.predict(X_test)
```

acuracia para os K

```python
from sklearn.metrics import accuracy_score
print( "acuracia do K=5 : ",accuracy_score(y_test,y_predicao5))
print( "acuracia do K=3 : ",accuracy_score(y_test,y_predicao3))
```

```python
plt.figure(figsize=(10,10))
plt.subplot(1,2,1)
plt.scatter(X_test[:,0],X_test[:,1], c=y_predicao5, marker='*',s=100,edgecolors='black')
plt.title("predio k com 5")

plt.subplot(1,2,2)
plt.scatter(X_test[:,0],X_test[:,1], c=y_predicao3, marker='*',s=100,edgecolors='black')
plt.title("predio k com 3")
plt.show()
```
